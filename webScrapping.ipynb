```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np 


company_names = []
company_address = []
industry_type = []



for i in range(1, 467):
        URL = f"https://www.vcsdata.com/hospitals-healthcare-in-india.php?page={i}&Industry=Hospitals/Healthcare"

        page = requests.get(url=URL, headers={
            "Content-Type": "text",
            "user-agent": "Mozilla/5.0 AppleWebKit/537.36 (KHTML, like Gecko; compatible; Googlebot/2.1; +http://www.google.com/bot.html) Chrome/W.X.Y.Z Safari/537.36"
        })

        soup = bs(page.text, "html.parser")

        sections = soup.findAll('section')

        div = sections[1].find_all(class_="col-xl-9 col-lg-9 col-md-12")

        div = bs(str(div), 'html.parser')

        cards = div.find_all(class_="card overflow-hidden text-darker font-weight-normal")

        for card in cards:
            name = card.find('h3').text
            company_names.append(name.replace('\n',""))
            address = card.find('span', attrs={"itemprop": "streetAddress"})
            address = (address.text).replace("Address : ", "")
            company_address.append(address)
            industry = card.find_all('div', class_="col-md-6")[1].text
            industry_type.append(industry.replace('Industry : ', ''))

        print("\nCompleted \n",i)

data_sheet = [company_names,industry_type,company_address]
data_sheet = np.array(data_sheet)
data_sheet = data_sheet.T
pd.DataFrame(data_sheet,columns=["Address","Type","Name"]).to_csv('./companyDetails.csv')
```
